A memória possui um barramento de endereçamento de 4 bits, o que permite 16 combinações possíveis. Cada endereço tem capacidade de armazenamento de 8 bits (1 byte), correspondendo à unidade de entrada de dados do sistema. O simulador permite gravar dados em um endereço específico, seja criando um novo registro ou substituindo um existente, além de realizar a leitura e a listagem de todos os endereços com suas respectivas funcionalidades.

```python

m0000 = "00000000"
m0001 = "00000000"
m0010 = "00000000"
m0011 = "00000000"
m0100 = "00000000"
m0101 = "00000000"
m0110 = "00000000"
m0111 = "00000000"
m1000 = "00000000"
m1001 = "00000000"
m1010 = "00000000"
m1011 = "00000000"
m1100 = "00000000"
m1101 = "00000000"
m1110 = "00000000"
m1111 = "00000000"

status = 1

while status == True:
    print("Opções: (W) Escrever / (R) Ler / (L) Listar / (0) Sair")
    opcao = input("Escolha uma opção: ")

    if opcao == '0':
        status = 0
        print("Encerrando simulador...")

    elif opcao == 'W' or opcao == 'w' or opcao == 'R' or opcao == 'r':
        escrevendo = 1
        while escrevendo == True:
            end = input("Digite o endereco de 4 bits: ")


            if len(end) != 4:
                print("ERRO: O endereço deve ter exatamente 4 bits.")
            else:

                if opcao == 'W' or opcao == 'w':
                    dado = input("Digite o dado de 8 bits: ")
                    if len(dado) != 8:
                        print("ERRO: O dado deve ter exatamente 8 bits.")
                    else:
                        if end == "0000": m0000 = dado
                        elif end == "0001": m0001 = dado
                        elif end == "0010": m0010 = dado
                        elif end == "0011": m0011 = dado
                        elif end == "0100": m0100 = dado
                        elif end == "0101": m0101 = dado
                        elif end == "0110": m0110 = dado
                        elif end == "0111": m0111 = dado
                        elif end == "1000": m1000 = dado
                        elif end == "1001": m1001 = dado
                        elif end == "1010": m1010 = dado
                        elif end == "1011": m1011 = dado
                        elif end == "1100": m1100 = dado
                        elif end == "1101": m1101 = dado
                        elif end == "1110": m1110 = dado
                        elif end == "1111": m1111 = dado
                        print("Gravado")


                else:
                    if end == "0000": print("Conteúdo:", m0000)
                    elif end == "0001": print("Conteúdo:", m0001)
                    elif end == "0010": print("Conteúdo:", m0010)
                    elif end == "0011": print("Conteúdo:", m0011)
                    elif end == "0100": print("Conteúdo:", m0100)
                    elif end == "0101": print("Conteúdo:", m0101)
                    elif end == "0110": print("Conteúdo:", m0110)
                    elif end == "0111": print("Conteúdo:", m0111)
                    elif end == "1000": print("Conteúdo:", m1000)
                    elif end == "1001": print("Conteúdo:", m1001)
                    elif end == "1010": print("Conteúdo:", m1010)
                    elif end == "1011": print("Conteúdo:", m1011)
                    elif end == "1100": print("Conteúdo:", m1100)
                    elif end == "1101": print("Conteúdo:", m1101)
                    elif end == "1110": print("Conteúdo:", m1110)
                    elif end == "1111": print("Conteúdo:", m1111)
                    else: print("Endereço não encontrado.")

            print("Digite W para escrever, R para ler ou outra tecla para voltar ao menu principal.")
            opcao = input()
            if opcao != 'W' and opcao != 'w' and opcao != 'R' and opcao != 'r':
                escrevendo = 0

    elif opcao == 'L' or opcao == 'l':
        print("Todos os estados:")
        print("0000: " + m0000)
        print("0001: " + m0001)
        print("0010: " + m0010)
        print("0011: " + m0011)
        print("0100: " + m0100)
        print("0101: " + m0101)
        print("0110: " + m0110)
        print("0111: " + m0111)
        print("1000: " + m1000)
        print("1001: " + m1001)
        print("1010: " + m1010)
        print("1011: " + m1011)
        print("1100: " + m1100)
        print("1101: " + m1101)
        print("1110: " + m1110)
        print("1111: " + m1111)

    else:
        print("Opcao invalida.")
```
