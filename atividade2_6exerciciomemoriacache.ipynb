#Exemplo 5.2:

Cálculo da quantidade de bits necessários para uma determinada memória cache.

Considere um sistema de computação com uma memória cache de 32KB de capacidade, constituída de linhas com 8 bytes de largura. A MP possui uma capacidade de 16MB.

##Resolução:
O total de bits a serem usados na cache é a soma dos bits de dados(produto da quantidade de linhas pela largura em bits de cada uma) mais os bits usados na coluna tag, cujo valor existe para cada linha. Então a fórmula é T(total de bits da cache) = Td(Total de bits para a parte de dados) + Tt(total de bits dos tags)

Capacidade Cache: 32 KB = 2¹⁵ bytes
Tamanho da Linha: 8 bytes = 2³ bytes
Capacidade MP: 16 MB = 2²⁴ bytes

O número de bits do endereço Total é 2²⁴ bytes, o endereço total possui 24 bits.

O offset é o tamanho da linha 2³, 3 bits para o offset.

Linhas = Cache/Linha = 2¹⁵/2³ = 2¹²

Tag = Endereço Total - Índice - Offset
Tag = 24 - 12 - 3 = 9 bits

Td = 2¹⁵.2³ = 2¹⁸

Tt = número de linhas * tamanho da tag
2¹² . 9 = 36.864 bits.
Total(T):
262.144 + 36.864 = 299.008 bits.
