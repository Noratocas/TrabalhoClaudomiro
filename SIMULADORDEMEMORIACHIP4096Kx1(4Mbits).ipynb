O programa grava, ler e lista simulando o funcionamento de uma memória Tannembaun. Nesse caso é uma memória com 1 bit de tamanho de conteúdo e 4.096K de capacidade de endereço.

image.png

```python
print("--- SIMULADOR DE MEMORIA CHIP 4.096K x 1 (4Mbits) ---")

memoria_log = ""
executando = True

while executando == True:
    print("W - Gravar / R - Ler / L - Listar / 0 - Sair")
    opcao = input("Escolha a opcao: ")

    if opcao == 'W' or opcao == 'w':
        ras = input("Digite a Linha (11 bits): ")
        cas = input("Digite a Coluna (11 bits): ")
        dado = input("Digite o Dado (1 bit): ")

        if len(ras) == 11 and len(cas) == 11 and len(dado) == 1 and (ras + cas + dado).replace('0', '').replace('1', '') == "":
            busca_completa = "RAS:" + ras + " CAS:" + cas
            nova_entrada = busca_completa + " Dado:" + dado


            if busca_completa in memoria_log:
                inicio_antigo = memoria_log.find(busca_completa)
                fim_antigo = memoria_log.find("#", inicio_antigo) + 1
                memoria_log = memoria_log[:inicio_antigo] + memoria_log[fim_antigo:]


            memoria_log = memoria_log + nova_entrada + "#"
            print("Gravado")
        else:
            print("ERRO: Tamanho invalido ou caractere nao binario!")

    elif opcao == 'R' or opcao == 'r':
        busca_r = input("Linha: ")
        busca_c = input("Coluna: ")
        busca_completa = "RAS:" + busca_r + " CAS:" + busca_c

        if busca_completa in memoria_log:
            inicio = memoria_log.find(busca_completa) + len(busca_completa) + 6
            print("Valor lido: " + memoria_log[inicio : inicio + 1])
        else:
            print("Endereco nao encontrado ou memoria vazia.")

    elif opcao == 'L' or opcao == 'l':
        if memoria_log == "":
            print("Memoria vazia.")
        else:
            print("--- Conteudo da Memoria ---")
            for entrada in memoria_log.split('#'):
                if entrada != "":
                    print(entrada)

    elif opcao == '0':
        print("Saindo...")
        executando = False

    else:
        print("Opcao invalida.")
```
