## Exemplo 4.1 Slide 72

Uma memória RAM(MP) tem um espaço máximo de endereçamento de 2K. Cada célula pode armazenar 16 bits. Qual o valor total de bits que podem ser armazenados nesta memória e qual o tamanho de cada endereço?

### Resolução:
A capacidade máxima do endereçamento é de 2K ou 2¹⁰.2¹ = 2¹¹ bits de endereço

Cada célula pode armazenar 16 bits ou 2⁴ bits de dados

Para calcular a capacidade que pode ser armazenado nessa memória é apenas multiplicar a capacidade do endereçamento pela tamanho da célula. Então:

2¹¹ . 2⁴ = 2¹⁵ bits ou 32Kbits

Já o tamanho de cada endereço é a quantidade do expoente em N resultante da capacidade máxima de endereçamento: 2¹¹, ou seja, o tamanho de cada endereço é 11 bits.

* 1 - (REM)<-------(CPU)

* 3 - (RDM) <---- (MP(REM))
* 4 - CPU <---- (RDM)

#Slide 88 - Questão 13

13) Considere uma memória com capacidade de armazenamento de 64K bytes; cada célula pode armazenar 1 byte de informação e cada caractere é codificado com 8 bits. Resolveu-se armazenar na memória deste sistema um conjunto de caracteres do seguinte modo. A partir do endereço (hexadecimal) 27FA, foram escritos sucessivamente grupos de 128 caracteres iguais, iniciando pelo grupo de As, seguido do grupo de Bs, e assim por diante. Qual deverá ser o endereço correspondente ao local onde está armazenado o 1. J?

##Resolução Questão 13:

Se começamos no endereço 27FA e foram escrito grupos de 128 seguindo a ordem alfabética, então até o primeiro J precisaria avançar no alfabeto 9 vezes e para passar os grupos precisa passar por todos os 128 caracteres. Então é só multiplicar 9 . 128 = 1.152 esse é a distância até o primeiro J, mas está em decimal e para fazer a soma precisamos transformar em hexadecimal. Então, 1.152 em hexadecimal é 480 e por fim é só fazer a soma do nosso endereço que é o ponto de partida 27FA e somar com 480 = 2C7A esse é o endereço do primeiro J.

14) O custo das memórias (SRAM) é maior que o das memórias (DRAM). No entanto, o processo de conexão das memórias (DRAM) é mais complexo que o das (SRAM) e, em consequência, o preço da interface das DRAM é bem maior que o das (SRAM). Supondo que uma interface de (DRAM) custe 5,00 reais, uma interface de (SRAM) custe 1,00 real, o preço por bit de uma SRAM é de 0,00002 reais e o de uma (DRAM) é de 0,00001 reais, calcule quantos bits deve ter uma memória dinâmica (DRAM) para que o conjunto seja mais barato.

##Resolução Questão 14:

Para encontrar o ponto onde a DRAM se torna a escolha economicamente viável, igualamos os dois custos:

1,00 + 0,00002n = 5,00 + 0,00001n

0,00002n-0,00001n = 5,00 - 1,00

0,00001n = 4,00

n = 4,0 / 0,00001 = n = 400.000 bits. Então aos 400.000 bits tanto a SRAM quanto DRAM tem o mesmo valor a partir desse valor para maior o DRAM é mais barato e se for um valor decrescente o SRAM é mais barato.
