O comportamento da cache ao receber dados:

```python
print("--- SIMULADOR DE CACHE ---")

tag00, dado00, v00 = "", "", False
tag01, dado01, v01 = "", "", False
tag10, dado10, v10 = "", "", False
tag11, dado11, v11 = "", "", False

executando = True

while executando == True:
    print("W - Escrever / R - Ler / L - Listar / 0 - Sair")
    opcao = input("Escolha a opcao: ")

    if opcao == 'W' or opcao == 'w':
        tag = input("Digite a Tag (2 bits): ")
        linha = input("Digite a Linha (2 bits): ")
        dado = input("Digite o Dado (1 bit): ")

        if len(tag) == 2 and len(linha) == 2 and len(dado) == 1:
            if linha == "00":
                tag00, dado00, v00 = tag, dado, True
            elif linha == "01":
                tag01, dado01, v01 = tag, dado, True
            elif linha == "10":
                tag10, dado10, v10 = tag, dado, True
            elif linha == "11":
                tag11, dado11, v11 = tag, dado, True
            print("Gravado")
            print("Linha: " + linha + " / Tag: " + tag + " / Dado: " + dado)
        else:
            print("ERRO: Formato invalido")

    elif opcao == 'R' or opcao == 'r':
        b_tag = input("Tag de busca (2 bits): ")
        b_linha = input("Linha de busca (2 bits): ")

        encontrado = False
        lido = ""

        if b_linha == "00" and v00 and tag00 == b_tag:
            encontrado, lido = True, dado00
        elif b_linha == "01" and v01 and tag01 == b_tag:
            encontrado, lido = True, dado01
        elif b_linha == "10" and v10 and tag10 == b_tag:
            encontrado, lido = True, dado10
        elif b_linha == "11" and v11 and tag11 == b_tag:
            encontrado, lido = True, dado11

        if encontrado == True:
            print("Resultado: Hit")
            print("Valor lido na Linha " + b_linha + ": " + lido)
        else:
            print("Resultado: Miss")
            print("Dado nao encontrado ou invalido para esta Tag.")

    elif opcao == 'L' or opcao == 'l':
        print("--- ESTADO DA CACHE ---")


        if v00:
            status00 = "Ocupada"
        else:
            status00 = "Vazia"
        print("Linha 00:  Tag:  " + tag00 + "  Dado: " + dado00 + "  Stat: " + status00 + "")

        if v01:
            status01 = "Ocupada"
        else:
            status01 = "Vazia"
        print("Linha 01:  Tag:  " + tag01 + "  Dado: " + dado01 + "  Stat: " + status01 + "")

        if v10:
            status10 = "Ocupada"
        else:
            status10 = "Vazia"
        print("Linha 10:  Tag:  " + tag10 + "  Dado: " + dado10 + "  Stat: " + status10 + "")

        if v11:
            status11 = "Ocupada"
        else:
            status11 = "Vazia"
        print("Linha 11:  Tag:  " + tag11 + "  Dado: " + dado11 + "  Stat: " + status11 + "")

    elif opcao == '0':
        print("Saindo...")
        executando = False
```
