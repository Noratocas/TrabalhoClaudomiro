O programa grava, ler e lista simulando o funcionamento de uma memória Tannembaun. Nesse caso é uma memória com 4 bits de tamanho de conteúdo e 128M de capacidade de endereço.

image.png

```python
print("--- SIMULADOR DE MEMORIA CHIP 128M x 4 (512Mbits) ---")

memoria_log = ""
executando = True

while executando == True:
    print("\nW - Gravar / R - Ler / L - Listar / 0 - Sair")
    opcao = input("Escolha a opcao: ")

    if opcao == 'W' or opcao == 'w':
        b = input("Banco (1 bit): ")
        r = input("Linha (13 bits): ")
        c = input("Coluna (13 bits): ")
        d = input("Dado (4 bits): ")

        if len(b) == 1 and len(r) == 13 and len(c) == 13 and len(d) == 4:
            busca_completa = "Banco:" + b + " RAS:" + r + " CAS:" + c
            nova_entrada = busca_completa + " Dado:" + d


            if busca_completa in memoria_log:
                inicio_antigo = memoria_log.find(busca_completa)
                fim_antigo = memoria_log.find("#", inicio_antigo) + 1
                memoria_log = memoria_log[:inicio_antigo] + memoria_log[fim_antigo:]

            memoria_log = memoria_log + nova_entrada + "#"
            print("Gravado")
        else:
            print("ERRO: Tamanho invalido.")

    elif opcao == 'R' or opcao == 'r':
        busca_b = input("Banco: ")
        busca_r = input("Linha: ")
        busca_c = input("Coluna: ")
        busca_completa = "Banco:" + busca_b + " RAS:" + busca_r + " CAS:" + busca_c

        if busca_completa in memoria_log:
            inicio = memoria_log.find(busca_completa) + len(busca_completa) + 6 # Pula ' Dado:'
            print("Valor lido: " + memoria_log[inicio : inicio + 4])
        else:
            print("Endereco nao encontrado.")

    elif opcao == 'L' or opcao == 'l':
        if memoria_log == "":
            print("Memoria vazia.")
        else:
            print("\n--- Conteudo da Memoria ---")
            for entrada in memoria_log.split('#'):
                if entrada != "":
                    print(entrada)

    elif opcao == '0':
        executando = False

    else:
        print("Opcao invalida.")
```
