Projete um sistema de memória que possui 13 pinos de endereços e um barramento de dados (BD) que transporta 2 palavras por vez. O BD não pode exceder 64 bits e a palavra deve ser maior que a célula da memória. A memória deve ter capacidade de 8 Gbits. Depois do projeto, indique qual é o tamanho da palavra, da célula, do BD e do barramento de endereços, e desenhe o chip.


# Projeto:
A capacidade da memória é de 8Gbits ou 2³⁰.2³=2³³

A capacidade de endereçamento é a relação da capacidade de memória por largura de barramento: 2³³/2⁶ = 2²⁷ ou 128M

O barramento de dados é o mesmo que a largura do barramento = sendo 64 bits ou 2⁶

O tamanho da palavra é 32 bits. Porque a memória transporta 2 palavras por vez com um barramento de largura limitado a 64 bits. Então 64/2=32.

O tamanho da célula é de 16 bits, mas poderia ser 8,4 ou 2. Desde que seja menor que a palavra de 32 bits, para questão de redundância manter o valor como base 2 se torna mais eficiente na transfêrencia de dados

Agora para o Barramento de endereços ele tem 13 pinos. Mas não é suficiente para suportar 2²⁷ da capacidade de endereçamento. Então, é utilizado RAS e CAS adicionando ao todo 26 barramento de endereço e para alcançar 2²⁷  e usa  1 linha de banco que corresponde a 2 bancos, porque as possibilidades de combinação são 2¹. Então: 2²⁶.2¹ = 2²⁷.
